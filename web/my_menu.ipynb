```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
url = 'https://www.menupan.com/restaurant/search/search_main_sub_area.asp?areacode=dj203'
html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
uls = soup.find(class_= 'areaList')
uls
```

```python
lis = uls.find_all(class_ = 'ifo')
lis[0]
```

```python
# 제목
lis[0].find('a').get_text()
```

```python
# 별점
float(lis[0].find(class_ = 'total').get_text())
```

```python
# 업종
lis[0].find_all('dd')[1].get_text().split('| ')[0]
```

```python
# 주소
lis[0].find_all('dd')[1].get_text().split('| ')[1]
```

```python
# 전화번호
lis[0].find_all('dd')[2].get_text()
```

```python
len(lis)
```

```python
title = []
score = []
category = []
address = []
phone = []

for li in lis:
    title.append(li.find('a').get_text())
    score.append(float(li.find(class_='total').get_text()))
    category.append(li.find_all('dd')[1].get_text().split('| ')[0])
    address.append(li.find_all('dd')[1].get_text().split('| ')[1])
    phone.append(li.find_all('dd')[2].get_text())
```

```python
import pandas as pd
my_menu = pd.DataFrame({'제목': title, '별점': score, '업종': category, '주소': address, '전화번호': phone})
my_menu
```

```python
my_menu.to_csv('data/my_menu.csv')
```
