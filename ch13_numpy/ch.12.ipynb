```python
# 12-2
plt.figure(figsize = (10, 5))
plt.plot(['kim', 'lee', 'kang'], [85, 88, 90])

plt.title('3명의 영어 성적')
plt.xlabel('학생 성적')
plt.ylabel('점수')

plt.show()
```

```python
# 12-4
xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
plt.plot(xdata, ydata1, label = '국어')
plt.plot(xdata, ydata2, label = '영어')
plt.legend(loc = 'upper center')
plt.title('세명 학생의 국어, 영어 성적')
plt.show()
```

```python
# 12-5 line style
xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
ydata3 = [85, 97, 78]
ydata4 = [92, 88, 82]
plt.figure(figsize = (10, 5))
plt.plot(xdata, ydata1, 'ro-', label = '국어', linewidth = 2, markersize = 10)
plt.plot(xdata, ydata2, 'cx--', label = '영어', linewidth = 3, markersize = 10)
plt.plot(xdata, ydata3, 'ms-.', label = '수학', linewidth = 4, markersize = 10)
plt.plot(xdata, ydata4, 'kd:', label = '사회', linewidth = 5, markersize = 10)

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc = 'best')
plt.show()
```

```python
# 12-6
xdata = list(range(30))
ydata = []

for x in xdata:
    y = 2 * x
    ydata.append(y)
plt.title('y = 2x')
plt.plot(xdata, ydata, color = 'magenta', linewidth = 3)
plt.ylim(0, 70)
plt.xlim(0, 35)
plt.xticks(list(range(0, 35, 2)))
plt.yticks(list(range(0, 70, 5)))

plt.grid(True)
plt.show()
```

```python
xdata = list(np.arange(0,10,0.0001))
sindata = np.sin(xdata)
cosdata = np.cos(xdata)
plt.plot(xdata, cosdata, label = 'cos')
plt.plot(xdata, sindata, label = 'sin')
plt.axhline(y = 0, color = 'k')
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.legend(loc = 'best')
plt.show()
```
