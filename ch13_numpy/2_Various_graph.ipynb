# Pie Chart

```python
# ex 12-9
pets = ['개', '고양이', '기타', '기르지않음']
portion = [18, 3.4, 3.1, 75.5]

plt.pie(portion, explode = (0, 0.1, 0, 0), labels = pets, autopct = '%1.1f%%', shadow = True, startangle= 0)
plt.title('총 가구 대비 반려동물 기르는 비율(2018)')
plt.show()
```

```python
data = list(range(1, 11))
min_temp = [7.3, 3.5, 7.2, 4.5, 3.5, 8.3, 7.7, 6.4, 9.3, 6.3]
max_temp = [15, 16.3, 15.2, 17.5, 18.3, 20.0, 18.8, 17.5, 14.2, 15.3]

plt.plot(data, min_temp, label = 'Min Temparature', color = 'red', linestyle = '-', marker = 'o')
plt.plot(data, max_temp, label = 'Max Temparature', color = 'blue', linestyle = '-', marker = 'x')
plt.legend(loc = 'best')
plt.xlabel('date')
plt.ylabel('temparature')
plt.title('Min and max temparature of ten days')
plt.show()
```

```python
xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
ydata3 = [85, 97, 78]
ydata4 = [92, 88, 82]

plt.plot(xdata, ydata1, label = '국어', color = 'red', linestyle = '-', marker = 'o')
plt.plot(xdata, ydata2, label = '영어', color = 'green', linestyle = '--', marker = 's')
plt.plot(xdata, ydata3, label = '수학', color = 'blue', linestyle = ':', marker = 'd')
plt.plot(xdata, ydata4, label = '사회', color = 'cyan', linestyle = '-.', marker = 'x')

plt.ylim(0, 100)
plt.title('세명 학생의 네 과목 성적')
plt.legend(loc = 'best')
plt.show()
```

## Contour (등고선) Plot

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title('Countour plots')
plt.contourf(XX, YY, ZZ, alpha = 0.75, cmap = 'jet')
plt.contour(XX, YY, ZZ, colors = 'k')
plt.show()
```
